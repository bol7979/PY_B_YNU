```python
from sklearn.preprocessing import LabelEncoder
```

```python
obj = {
    "Class": ["A", "B", "A", "C", "E", "D", "D", "A", "F"]
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(obj)
df
```

```python
encoder = LabelEncoder()
```

```python
encoder.fit(df.Class) #df["Class"]
```

```python
labels = encoder.transform(df.Class)
labels
```

```python
df["to_수치형데이터"] = labels
df
```

```python
obj2 = {
    "과일": ["사과", "포도", "수박", "딸기", "배", "사과", "사과", "수박"]
}

df2 = pd.DataFrame(obj2)
df2
```

```python
encoder = LabelEncoder()
encoder.fit(df2) #학습
labels2 = encoder.transform(df2) #변형

df2["to_수치데이터"] = labels2
df2
```

```python
import numpy as np

conditionList = [
    (df["Class"] == "A"),
    (df["Class"] == "B"),
    (df["Class"] == "C"),
    (df["Class"] == "D"),
    (df["Class"] == "E"),
    (df["Class"] == "F")
]

choiceList = list(range(6))

np_labels = np.select(condlist=conditionList, choicelist=choiceList)

df["Numpy_수치형데이터"] = np_labels
df
```
